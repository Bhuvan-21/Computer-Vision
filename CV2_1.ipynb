```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline


img = cv2.imread("Images/New_York_City_.png")           # reading an image


plt.figure(figsize = (5, 5))
plt.imshow(img)
```

```python
cv2.imwrite("out.jpg", img)                 # saving an image to the disk
```

```python
black = np.ones([150, 200, 3], 'uint8') * 0
grey = np.ones([150, 200, 3], 'uint8') * 128
white = np.ones([150, 200, 3], 'uint8') * 255
img = [black, grey, white]
labels = ["black", "grey", "white"]
fig = plt.figure(figsize = (25, 5))
for i in range(3):
    ax = fig.add_subplot(1, 3, i+1)
    plt.imshow(img[i])
    plt.title(labels[i])
plt.show()
```

```python
img = cv2.imread("Images/eYZY4l-9Ibo0UV5x_wr-Gg.jpg")


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # converting to gray scale
plt.figure(figsize = (5, 5))
plt.imshow(gray, 'gray')
```

```python
#viewing diff colour channels
bgr = cv2.imread("Images/2009-12-06 06_58_39.jpg")
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
r, g, b = rgb[:, :, 0], rgb[:, :, 1], rgb[:, :, 2]    

img = [rgb, r, g, b]
labels = ["Original", "Red", "Green", "Blue"]
fig = plt.figure(figsize = (25, 5))
for i in range(4):
    ax = fig.add_subplot(1, 4, i+1)
    if i != 0:
        plt.imshow(img[i], 'gray')
    else:
        plt.imshow(img[i])
    plt.title(labels[i])
plt.show()
```

```python
img = cv2.imread("Images/2009-12-06 06_58_39.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_copy = img.copy()

cv2.rectangle(img_copy, pt1 = (40, 200), pt2 = (60, 250), 
              color = (255, 0, 0), thickness = 2)

cv2.circle(img_copy, center = (125, 125), radius = 25, 
                       color = (87, 184, 237), thickness = 2)

cv2.line(img_copy, pt1 = (125, 150), pt2 = (125, 225), color = (255, 0, 0), thickness = 2)
cv2.line(img_copy, pt1 = (125, 175), pt2 = (85, 200), color = (255, 0, 0), thickness = 2)
cv2.line(img_copy, pt1 = (125, 175), pt2 = (165, 200), color = (255, 0, 0), thickness = 2)
cv2.line(img_copy, pt1 = (125, 225), pt2 = (105, 250), color = (255, 0, 0), thickness = 2)
cv2.line(img_copy, pt1 = (125, 225), pt2 = (145, 250), color = (255, 0, 0), thickness = 2)

plt.imshow(img_copy)
```
